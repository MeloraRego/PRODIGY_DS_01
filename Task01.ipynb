```python
# Bar Chart for Income Distributions by Regions
income_by_region = pd.crosstab(merged_table.IncomeGroup,merged_table.Region)
plt.figure(figsize=(10, 6))
income_by_region.plot(kind='bar')
plt.xlabel("Income Group")
plt.ylabel("Number of Countries in the Regions")
plt.title("Income Distribution by Regions")
plt.legend(title='Regions')
plt.xticks(rotation=45)
plt.show()
```

```python
income_group_counts = merged_table['IncomeGroup'].value_counts() #getting count of different income group
print(income_group_counts)
```

```python
income_group_counts.plot(kind='bar', color='gray') #plotting
plt.xlabel('Income Group')
plt.ylabel('Count')
plt.title('Counts of Income Groups')
plt.show()
```

Interpretation for the above:
Among the 215 countries 82 countries have high income, 54 countries belong to lower middle income, 53 countries belong to upper middle income and 26 countries belong to low income

```python
columns_01 = [2012.0, 2013.0, 2014.0, 2015.0, 2016.0, 2017.0, 2018.0, 2019.0, 2020.0, 2021.0, 2022.0] #create new df for storing years from 2012-2022
```

```python
#calcualte the population growth rate 
merged_table.loc[:,'Population_Growth_Rate'] = ((merged_table[2022.0] - merged_table[2012.0]) / merged_table[2012.0]) * 100
```

```python
# Aggregate by income group and calculate average growth rates
average_growth_rates = merged_table.groupby('IncomeGroup')['Population_Growth_Rate'].mean().reset_index()
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.bar(average_growth_rates['IncomeGroup'], average_growth_rates['Population_Growth_Rate'], color='pink')
plt.xlabel('Income Group')
plt.ylabel('Average Population Growth Rate (%)')
plt.title('Average Population Growth Rates by Income Group (2012-2022)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Interpretation for the above: according to the facts provided countries with higher income have lesser average population growth rate and the inverse for the low income countries which are leading in the average population growth rate, it suggests that countries tend to experience a decline in birth rates or increase in death, migrants and population growth as they develop economically. 

Displaying countries with low income

```python
#create a df and consider only countries with low income
low_income_countries = merged_table[merged_table['IncomeGroup'] == 'Low income']
```

```python
selected_columns = ['Country Name', 2012.0, 2013.0, 2014.0, 2015.0, 2016.0, 2017.0, 2018.0, 2019.0, 2020.0, 2021.0, 2022.0]
```

```python
new_df
```

```python
population_columns = [2012.0, 2013.0, 2014.0, 2015.0, 2016.0, 2017.0, 2018.0, 2019.0, 2020.0, 2021.0, 2022.0]
```

```python
# Calculating the row-wise average
new_df['Average Population'] = new_df[population_columns].mean(axis=1)
```

```python
# Displaying the DataFrame with average population
print(new_df[['Country Name', 'Average Population']])
```

```python
plt.figure(figsize=(12, 8))
plt.bar(new_df['Country Name'], new_df['Average Population'], color='gray')
plt.xlabel('Country')
plt.ylabel('Average Population')
plt.title('Average Population Across Years (2012-2022) with low income')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Interpretation for the above: The countries like Ethiopia,congo,Dem.Rep, Afghanistan, Sudan,uganda etc have comparatively more number of average population with the lowest income these could be many unknown causes like limited resources, standard of living, political instablity etc.
